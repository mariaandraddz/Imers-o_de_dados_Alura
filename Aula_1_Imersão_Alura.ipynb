```python
import numpy as np
dataframe_salarios = pd.DataFrame({
    "nome" : ['joao', 'ana', 'dani'],
    "salario" : [1000, np.nan, 3000]
})

dataframe_salarios['salario_media'] = dataframe_salarios['salario'].fillna(dataframe_salarios['salario'].mean().round(2)) #dataframe_salarios é a base de dados; salario_media é a nova coluna que esta sendo criada; dentro da base os lugares nulos com a media e com valores arredondados
print(dataframe_salarios)
```

```python
# a ana estava sem salario, a partir da media que realizamos ela passou a ter 2000. A media = 1000 + 3000/2 = 2000 (por isso deu esse valor)
```

```python
import numpy as np
dataframe_salarios = pd.DataFrame({
    "nome" : ['joao', 'ana', 'dani'],
    "salario" : [1000, np.nan, 3000]
})

dataframe_salarios['salario_mediana'] = dataframe_salarios['salario'].fillna(dataframe_salarios['salario'].median())
print(dataframe_salarios)
```

```python
dataframe_temperaturas = pd.DataFrame({
    'dia' : ['segunda', 'terça', 'quarta', 'quinta', 'sexta', 'sabado', 'domingo'],
    'temperatura' : [30, np.nan, np.nan, 33, 32, 31, 29]
})

dataframe_temperaturas['preenchido_fill'] = dataframe_temperaturas['temperatura'].ffill()
print(dataframe_temperaturas)

#ffill: completa com o valor anterior, por isso terça e quarta que eram valores nullos receberam os mesmos valores que segunda
```

```python
dataframe_temperaturas = pd.DataFrame({
    'dia' : ['segunda', 'terça', 'quarta', 'quinta', 'sexta', 'sabado', 'domingo'],
    'temperatura' : [30, np.nan, np.nan, 33, 32, 31, 29]
})

dataframe_temperaturas['preenchido_bfill'] = dataframe_temperaturas['temperatura'].bfill()
print(dataframe_temperaturas)
#bfill: pega do valor posterior
```

```python
dataframe_cidades = pd.DataFrame({
    'nome' : ['joao', 'ana', 'dani'],
    'cidades' : ['São Paulo', np.nan, 'Curitiba']
})

dataframe_cidades['cidades_preenchidas'] = dataframe_cidades['cidades'].fillna('Desconhecido')
print(dataframe_cidades)
```

# AULA 3 - VISUALIZAÇÃO DE DADOS

```python
dataframe_limpo.head()
```

```python
dataframe_limpo['senioridade'].value_counts().plot(kind='bar', title='Distribuição de Senioridade', color='red')
```

```python
import seaborn as sns #mexer com graficos de forma mais aprofundado
import matplotlib.pyplot as plt
```

```python
sns.barplot(data=dataframe_limpo, x='senioridade', y='salario_em_usd')
```

```python
plt.figure(figsize=(8, 5)) #tamanho da figura
sns.barplot(data=dataframe_limpo, x='senioridade', y='salario_em_usd')
plt.title('Salário médio por senioridade')
plt.xlabel('Senioridade')
plt.ylabel('Salário em anual USD')
plt.show()
```

```python
dataframe_limpo.groupby('senioridade')['salario_em_usd'].mean().sort_values(ascending=False)
```

```python
#ordenar do maior para o maior(false) ou o contrário
ordem = dataframe_limpo.groupby('senioridade')['salario_em_usd'].mean().sort_values(ascending=False).index
print(ordem)
```

```python
plt.figure(figsize=(8, 4)) #tamanho da figura
sns.barplot(data=dataframe_limpo, x='senioridade', y='salario_em_usd', color='red', order=ordem)
plt.title('Salário médio por senioridade')
plt.xlabel('Senioridade')
plt.ylabel('Salário em anual USD')
plt.show()
```

```python
plt.figure(figsize=(8, 5)) #tamanho
sns.histplot(dataframe_limpo['salario_em_usd'], bins = 40, kde=True, color='red')
plt.title('Distribuição de Salários anuais')
plt.xlabel('Salario anual USD')
plt.ylabel('Frequência')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=dataframe_limpo, x=dataframe_limpo['salario_em_usd'])
plt.title('Boxplot Distribuição de Salários anuais')
plt.ylabel('Frequência')
plt.show()
```

```python
ordem_senioridade = ['Junior', 'Pleno', 'Senior', 'Executivo']
plt.figure(figsize=(8, 5))
sns.boxplot(x='senioridade', y='salario_em_usd', data=dataframe_limpo, order= ordem_senioridade, palette='Set2', hue='senioridade')
plt.title('Boxplot Distribuição por senioridade')
plt.xlabel('Senioridade')
plt.ylabel('Salário em anual USD')
plt.show()
```

```python
import plotly.express as px #gráficos interativos

salario_medio_por_senioridade = dataframe_limpo.groupby('senioridade')['salario_em_usd'].mean().reset_index()

fig = px.bar(
    salario_medio_por_senioridade,
    x='senioridade',
    y='salario_em_usd',
    title='Salário Médio por Senioridade',
    labels={'senioridade': 'Senioridade', 'salario_em_usd': 'Salário Médio Anual (USD)'},
    color='senioridade',
    template='plotly_white'
)

fig.update_layout(xaxis_title='Senioridade', yaxis_title='Salário Médio Anual (USD)')
fig.show()
```

```python
contagem_senioridade = dataframe_limpo['senioridade'].value_counts().reset_index()
cargos_contagens_columns = ['senioridade', 'contagem']
contagem_senioridade.columns = cargos_contagens_columns
fig = px.pie(
    salario_medio_por_senioridade,
    title='Salário Médio por Senioridade',
    labels={'senioridade': 'Senioridade', 'salario_em_usd': 'Salário Médio Anual (USD)'},
    color='senioridade',
    template='plotly_white'
)

fig.update_layout(xaxis_title='Senioridade', yaxis_title='Salário Médio Anual (USD)')
fig.show()
```
